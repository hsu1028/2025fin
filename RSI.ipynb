```python
no = len(df)
```

```python
buy_list =[]
sell_list = []
for i in range(0,no):
    buy = 0
    sell = 0
    if RSI[i]>70:
        sell=1
    if RSI[i]<30:
        buy=1
    buy_list.append(buy)
    sell_list.append(sell)
```

```python
import pandas as pd
RSI_df = pd.DataFrame(RSI, columns=["RSI"], index=df.index)
```

```python
buy_df = pd.DataFrame(buy_list, columns=["Buy_Signal_RSI"], index=df.index)
```

```python
sell_df = pd.DataFrame(sell_list, columns=["Sell_Signal_RSI"], index=df.index)
```

```python
df_merged = pd.concat([df, RSI_df, buy_df, sell_df], axis=1, join="inner")
```
